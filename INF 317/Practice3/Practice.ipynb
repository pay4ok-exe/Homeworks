```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('country_wise_latest.csv')
data.head()
```

```python
top_countries = data.nlargest(10, 'Confirmed')
```

```python
plt.figure(figsize=(10, 6))
top_countries.plot(kind='bar', x='Country/Region', y='Confirmed', color='red', legend=False)
plt.title('Top 10 Countries by Confirmed COVID-19 Cases')
plt.xlabel('Country')
plt.xticks(rotation=-45)
plt.tight_layout()
plt.show()
```

```python
labels = ['Confirmed', 'Recovered', 'Deaths']
sizes = [top_countries['Confirmed'].sum(), top_countries['Recovered'].sum(), top_countries['Deaths'].sum()]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightcoral', 'darkblue'])
plt.title('COVID-19 Cases Distribution for Top 10 Countries')
plt.axis('equal')
plt.show()
```
